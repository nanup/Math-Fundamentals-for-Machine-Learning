```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 10000)
```

```python
def f(my_x):
  my_y = my_x**2 + 2*my_x + 2
  return my_y
```

```python
y = f(x)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
_ = ax.plot(x,y)
```

```python
f(2)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')
plt.scatter(2,10)
_ = ax.plot(x,y)
```

```python
f(5)
```

```python
def diff_demo(my_f, my_x, my_delta):
  return(my_f(my_x + my_delta) - my_f(my_x)) / my_delta
```

```python
deltas = [1, 0.1, 0.001, 0.001, 0.00001, 0.000001]
```

```python
for delta in deltas:
  print(diff_demo(f, 2, delta))
```

```python
for delta in deltas:
  print(diff_demo(f, -1, delta))
```
